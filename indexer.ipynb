```python
import re 
import sys
from bs4 import BeautifulSoup
unique_words=set()

for file in os.listdir(directory):  #The "os" module iterates through all the files and directories in the "directory" 
    file =  os.path.join(directory, file) #to join the directory and file name to create the full path to the file.
    f=open(file, encoding='utf-8') #to ensure that the file is read correctly.
    text = f.read()
    soup = BeautifulSoup(text, 'lxml') #to parse the contents of the file and extract text.
    text = soup.get_text()
    # each string in the resulting list represents a token that consists of one or more Urdu characters using re.findall.
    urdu_text = re.findall(r'[\u0600-\u06FF\u200c]+', text) #to extract Urdu text from the text extracted from the file.
    unique_words.update(urdu_text)
    f.close()

print(unique_words)
```

```python
stopwords = set()
with open("Urdu stopwords.txt",encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        stopwords.add(str(line))


#Loop through the document and check if each token is in the stop_words list
cleaned_unique_words = set()
for token in list(unique_words):
    if token not in stopwords:
        cleaned_unique_words.add(token)

#Output the document without any stop words
```

```python
doc_ids = dict()
import os
path = "Documents"
allfiles = os.listdir(directory)
docid = 1
# Open a file to write the output
#This code writes the file paths of all files in a directory to a file called "docids.txt".
# It also assigns a unique ID to each file and stores the ID in a dictionary called "doc_ids"
with open('docids.txt', 'w', encoding= 'utf-8') as o:
    for f in allfiles:
        o.write('{}\{}\n'.format(docid, f))
        doc_ids[f] = docid
        docid += 1
```

```python
#This code creates a dictionary of term IDs (term_ids) for a list of cleaned and unique words
# and then writes the term IDs and corresponding words to a file called "termids.txt".

term_ids = dict()
with open('termids.txt', 'w',encoding='utf-8') as f:
    termid = 0
    for token in list(cleaned_unique_words):
        term_ids[token] = termid
        f.write(f"{termid}\{token}\n")
        termid += 1
```

**Forward Index**

A forward index is an index that maps document IDs to the terms

doc1: term1->term2

```python
file_names = os.listdir(directory)
forward_index = dict()
for f_name in file_names: #initializes an empty dictionary called forward_index to store the forward index
    file_path = os.path.join(directory, f_name) #It also retrieves the unique ID of the file from the doc_ids dictionary using the filename (f_name) as the key
    f_id = doc_ids[f_name]
    forward_index[f_id] = dict()

    with open(file_path, 'r', encoding  = 'utf-8') as f1:
        words = f1.read().split()
        for i in range(len(words)):
            if words[i] in cleaned_unique_words:
                word_id = term_ids[words[i]] #The code then checks if the word_id key already exists in the forward_index[f_id] dictionary
                if forward_index[f_id].get(word_id) == None:#the code creates a new list containing the current index (i) 
                    forward_index[f_id][word_id] = list() # and assigns it to forward_index[f_id][word_id]
                    forward_index[f_id][word_id].append(i)
                else:
                    forward_index[f_id][word_id].append(i) #the code appends the current index to the existing list here.
```

```python
# Open a file to write the forward index
files_names = os.listdir(directory)

with open("forwardIndex.txt","w",encoding='utf-8') as f1:
    # Iterate over all documents
    for doc_id in forward_index.keys():
        # Iterate over all the tokens in each document
        for term_id in forward_index[doc_id].keys():
            # Get the positions of the current token in the document
            positions = forward_index[doc_id][term_id]
            # Write the DOCID, TERMID, and positions to the file
            f1.write(f'{doc_id}')
            f1.write("\{}".format('t'))
            f1.write(f'{term_id}')
            for v in positions:
                 f1.write("\{}".format('t'))
                 f1.write(f'{v}')
            f1.write(f'\n')
```

the forward index file will be in forwardindex.txt with format
DOCID, TERMID, and positions


**Inverted index**

an index that maps for each term in a doc

term1: doc1->doc2->doc8


```python
global inverted_index
inverted_index = {} #it will store the inverted index

#read the forward_index.txt file and store it in a dictionary
with open("doc_index.txt", "r") as f:
    for line in f:
        line = line.split()
        file = line[0] #. The first component (file) is the ID of the document
        term_id = int(line[1]) #the second component (term_id) is the ID of the term
        positions = [int(x) for x in line[2:]]
        if term_id not in inverted_index:
            inverted_index[term_id] = {}
        inverted_index[term_id][file] = positions  #the remaining components (positions) are the positions of the term within the document


previous_key=0
#store the inverted index in a txt file in append mode
with open("term_index.txt", "a") as f:
    #delta encode when you subtract the previous document ID from the current document ID before writing it to the file
    for key, value in inverted_index.items():
        f.write(str(key) + " ") #the code writes the term_id to the file followed by a space
        #then store the file name and the position
        previous_key=0
        for key1, value1 in value.items():
            f.write(str(int(key1)-int(previous_key)) + ":")#The document IDs are stored as the difference between the current document ID 
            previous_key=key1#and the previous document ID, in order to save space.
            previous_i=0
            for i in value1:
                f.write(str(i-previous_i) + " ")
                previous_i=i
#This reduces the number of bits required to store the document IDs, and can therefore result in a smaller index file size
        f.write("\n")


doc1=0
temp={} #will store the number of occurrences of each term
doc1={}# the number of documents in which each term appear
pos1={} # the position of each term's information in the term_info.txt file

#read the inverted_index.txt file and store the number of occurence of each term in a dictionary temp and number of doucments in a variable
with open("term_index.txt", "r") as f:
    length=0

    for line in f:
        a = line.split()
        #if there is a : in the line then increment the number of documents
        term_id = int(a[0]) #The first component (term_id) is the ID of the term
        pos1[term_id]=length
        
#the code increments the number of occurrences of the term in the temp dictionary 
#and increments the number of documents in which the term appears in the doc1 dictionary. 
#The code also stores the position of the term's information in the term_info.txt file in the pos1 dictionary.

        temp[term_id]=len(a)-1
        
        b=line.split(":")
        doc1[term_id]=len(b)-1 

        length=len(line)+length
        

#store the number of occurence of each term in a txt file in append mode
with open("term_info.txt", "a") as f:
    for key, value in temp.items():
        f.write(str(key) + "/" + str(pos1[key]) +"/" + str(value)+"/"+ str(doc1[key]) + "\n")
        #The format for each line is term_id/position_in_file/number_of_occurrences/number_of_documents.
        # The position of each term's information in the file is obtained from the pos1 dictionary.
```
